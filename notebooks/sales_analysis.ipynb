```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv('/content/superstore_cleaned.csv')
df.head()
```

```python
df.describe()
```

```python
df.nunique()
```

```python
print("Total Sales:", round(df['Sales'].sum(), 2))
print("Total Profit:", round(df['Profit'].sum(), 2))
print("Total Orders:", df['Order ID'].nunique())
```

```python
sales_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=sales_region.index, y=sales_region.values)
plt.title('Total Sales by Region')
plt.ylabel('Sales')
plt.show()
```

```python
monthly_sales = df.groupby('Order Month')['Sales'].sum().reindex([
    'January','February','March','April','May','June','July','August','September','October','November','December'
])

sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
sns.barplot(x=category_profit.index, y=category_profit.values)
plt.title('Total Profit by Category')
plt.ylabel('Profit')
plt.show()
```

```python
# Create the aggregation (Step 1: Calculate Sales by Product)
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)

# You can check the result before plotting
print(top_products)
```

```python
# Step 2: Plot the results
sns.barplot(x=top_products.values, y=top_products.index) # Use values for x (sales), index for y (product names)
plt.title('Top 5 Products by Sales')
plt.xlabel('Sales')
plt.ylabel('Product Name') # Add a label for the Y-axis (Product Name)
plt.show()
```

```python
# Calculate the difference between Ship Date and Order Date
df['Shipping Time'] = df['Ship Date'] - df['Order Date']

# Convert the time difference (timedelta) into days (integer)
df['Shipping Time (Days)'] = df['Shipping Time'].dt.days

# Check the new column
print(df[['Order Date', 'Ship Date', 'Shipping Time (Days)']].head())
```

```python
# Now the column exists, this code will work
sns.histplot(df['Shipping Time (Days)'], bins=20, kde=True)
plt.title('Distribution of Shipping Time')
plt.xlabel('Days')
plt.ylabel('Count of Orders') # Good practice to specify the y-label
plt.show()
```
